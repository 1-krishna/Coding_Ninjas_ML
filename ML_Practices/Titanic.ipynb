```python
#Dropping name column
df.drop("Name", axis = 1, inplace = True)
```

```python
#changing gender to int
#0 for female and 1 for male

def change_gender(gen):
    if(gen=="male"):
        return 1
    else:
        return 0

df["Sex"] = df.Sex.apply(change_gender)
df.head()
```

```python
df.isnull().sum()
```

```python
#converting age to integer

def change_to_int(age):
    return int(age)

df["Age"] = df.Age.apply(change_to_int)
df.head()
```

```python
#dropping passengerId column
df.drop("PassengerId", axis = 1, inplace = True)
df.head()
```
