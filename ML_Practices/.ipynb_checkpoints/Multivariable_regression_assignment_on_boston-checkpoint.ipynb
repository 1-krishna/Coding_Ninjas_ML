```python
from sklearn import datasets
import pandas as pd
import numpy as np
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
```

```python
boston_data = datasets.load_boston()
boston_data
```

```python
X = boston_data.data
Y = boston_data.target

df = pd.DataFrame(X)
df.columns = boston_data.feature_names
df.describe()

df2 = df.copy()

for i in range (len(df.columns)):
    new_col = df[df.columns[i]] ** 2
    df2[df.columns[i]+"_"+df.columns[i]] = new_col
#df2.describe()

for i in range (len(df.columns)):
    for j in range(i+1, len(df.columns)):
        new_row = df[df.columns[i]] * df[df.columns[j]]
        df2[df.columns[i]+"_"+df.columns[j]] = new_row
df2.describe()
```

```python
alg1 = LinearRegression()
X_new = df2.values

x_train, x_test, y_train, y_test = model_selection.train_test_split(X_new, Y)

alg1.fit(x_train, y_train)
```

```python
train_score = alg1.score(x_train, y_train)
test_score = alg1.score(x_test, y_test)

print("train score ", train_score)
print("test score ", test_score)

#it was just going upto .77 in one degree linear regression
```
