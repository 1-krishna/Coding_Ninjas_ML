#String based data handling

```python
df["Gender"] = "Female"
df.iloc[0:10, 6] = "Male"
df.head(15)
```

```python
#to string data to some other type
def f(s):
    if s == "Male":
        return 0
    else:
        return 1

df["sex"] = df.Gender.apply(f)
df.head()

#now we can delete gender col
del df["Gender"]
df.head()
```
