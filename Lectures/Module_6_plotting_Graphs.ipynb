```python
#Using MATPLOTLIB

import matplotlib.pyplot as plt
```

```python
#Just plutting dots 
x = [1,2,3]
y = [2,4,6]
plt.scatter(x, y)
plt.show()
```

```python
#for line connecting these
x = [1,2,3]
y = [2,4,6]
plt.plot(x, y)
plt.show()
```

```python
plt.scatter(x, y)
plt.plot(x, y)
plt.show()
```

```python
#two different lines
x = [1,2,3]
y = [2,4,6]

x2 = [4,5,6]
y2 = [7,8,9]

plt.plot(x, y)
plt.plot(x2, y2)
plt.show()
```

```python
#deciding own color
plt.plot(x, y, "r-") #first character decides color
                    #second character decides type of line
#plt.plot(x2, y2, color="blue", marker="--")
plt.plot(x2, y2, "b--")
# similarly +, p(for pentagons), o, ^, -., etc
plt.show()

#by using linewidth argument we could set the width of line also
```

```python
#plotting function x^3
import numpy as np
x = np.array([1,2,3,4])
y = x**3
plt.plot(x,y)
plt.show()
```

```python
#for more precise points
x = np.arange(0,5,0.1)
y = x**3
plt.plot(x, y)
plt.show()
```

```python
#if we have n points and we are passing only one array
#then it will automatically take 1----n as x
#e.g
a = [1,2,3]
plt.plot(a)
plt.show()
```

```python
"""
color, marker, linewidth, labels for axis,
datasets and the title, axis, grid, text
"""
x = np.arange(0,5,0.1)
y = x**3

plt.plot(x,y,color="blue", linewidth=5, label="x^3")

y2 = x**2

plt.plot(x,y2, color="red", linewidth=5, label="x^2")
#setting labels

#manually setting axis
plt.axis([0,5,0,150])#4args xmin,xmax,ymin,ymax
plt.xlabel("x")
plt.ylabel("x^3")

#to add grids
plt.grid()
plt.legend()
plt.title("Matplot lib demo session")

#to add a text at some x,y

plt.text(2,80, "Text", fontsize=12)


plt.show()
```

```python
#pie graphs
sizes = [2,4,6,2]
plt.pie(sizes)
plt.show()
```

```python
#to make it round
sizes = [2,4,6,2]
plt.pie(sizes)
plt.axis("equal")
plt.show()
```

```python
#changing colors
sizes = [2,4,6,2]
colors = ["blue", "red", "green", "purple"]
plt.pie(sizes, colors = colors)
plt.axis("equal")
plt.show()
```

```python
#adding labels
sizes = [2,4,6,2]
colors = ["blue", "red", "green", "purple"]
labels = ["A", "B", "C", "D"]
plt.pie(sizes, colors = colors, labels=labels)
plt.axis("equal")
plt.show()
```

```python
#adding data in terms of percentage
#using autopct
#in autopct we can also pass a function

sizes = [2,4,6,2]
colors = ["blue", "red", "green", "purple"]
labels = ["A", "B", "C", "D"]
plt.pie(sizes, colors = colors, labels=labels, autopct="%.1f")
plt.axis("equal")
plt.show()
```

```python
#adding % sign at last
sizes = [2,4,6,2]
colors = ["blue", "red", "green", "purple"]
labels = ["A", "B", "C", "D"]
plt.pie(sizes, colors = colors, labels=labels, autopct="%.1f%%")
plt.axis("equal")
plt.show()
```

```python
#to explode(Cut a piece)

sizes = [2,4,6,2]
colors = ["blue", "red", "green", "purple"]
labels = ["A", "B", "C", "D"]
explode = [0.1,0,0,0]
plt.pie(sizes, colors = colors, explode = explode, labels=labels, autopct="%.1f%%")
plt.axis("equal")
plt.show()
```

```python
#to change direction from ACK to Clockwise
#adding % sign at last
sizes = [2,4,6,2]
colors = ["blue", "red", "green", "purple"]
labels = ["A", "B", "C", "D"]
explode = [0.1,0,0,0]
plt.pie(sizes, colors = colors,explode=explode, counterclock=False, labels=labels, autopct="%.1f%%")
plt.axis("equal")
plt.show()
```

```python
#Setting startAngle
sizes = [2,4,6,2]
colors = ["blue", "red", "green", "purple"]
labels = ["A", "B", "C", "D"]
explode = [0.1,0,0,0]
plt.pie(sizes, colors = colors,startangle=100,explode=explode, counterclock=False, labels=labels, autopct="%.1f%%")
plt.axis("equal")
plt.show()
```
